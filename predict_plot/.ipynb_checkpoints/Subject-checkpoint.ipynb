```python
with open('test.txt') as files:
    output = {}
    for line in files.readlines():
        line = line.strip()
        if line.count(',') == 1:
            output[line] = []
            currentPoint = line
        else:
            temp = list(map(float, line.split(',')))
            output[currentPoint].append(temp)
```

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
```

```python
def convert_to_cm(x, y):
    dWidth = 58.5
    dHeight = 104.05
    screenWidth = 375
    screenHeight = 667
#             let dX = 18.61
#         let dY = 8.03
    dx = 18.61
    dy = 8.03
    x = x/screenWidth*dWidth
    y = y/screenHeight*dHeight
    x -= dx
    y = y + dy
    return (x/10, y/10)

convert_to_cm(1, 1)

def euclidean_distance(x1, y1, x2, y2):
    return np.sqrt( (x1-x2)**2 + (y1-y2)**2)
```

```python
count = 0
for index, key in enumerate(output):
    if(len(output[key]) < 5):
        continue
    fig, axes = plt.subplots()
    axes.set_xlim(0, 375)
    axes.set_ylim(-675, 0)
    x, y = list(map(float, key.split(',')))
    y = -y
    print(x,y)
    axes.scatter(x, y, color='red')
    x1, y1 = convert_to_cm(x, y)
    for i in output[key]:
        pre_x, pre_y = i[2], -i[3]
        axes.scatter(pre_x, pre_y, color='blue')
        x2, y2 = convert_to_cm(pre_x, pre_y)
```

```python
count = 0
total = 0
total_out_screen
for index, key in enumerate(output):
    if(len(output[key]) < 5):
        continue
    x, y = list(map(float, key.split(',')))
    x1, y1 = convert_to_cm(x, y)
    for i in output[key]:
        count += 1
        pre_x, pre_y = i[2], i[3]
        print(pre_x, pre_y)
        if pre_x >= 0 and pre_x <= 375 and pre_y >= 0 and pre_y <= 675:
            x2, y2 = convert_to_cm(pre_x, pre_y)
    #         print(x2, i[0])
    #         print(x2, y2)
            total += euclidean_distance(x1, y1, x2, y2)
    #         print(euclidean_distance(x1, y1, x2, y2))
        else:
            total_out_screen += 1
print(total)
print(count)
print(total/count)
```
